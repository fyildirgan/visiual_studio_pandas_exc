```python
import os
```

```python
folder_path = './image'
paths = os.listdir(folder_path)
```

```python

for path in paths:
    if path.endswith('.jpg'):
        img = Image.open(os.path.join(folder_path, path))
        thumb = img.copy()
        thumb.thumbnail((300, 300))
        thumb = thumb.transpose(Image.FLIP_LEFT_RIGHT)
        thumb.save(f'./process_image/{path}')


```
